```python
import requests
```

```python
url = 'https://www.sec.gov/Archives/edgar/data/320193/000032019317000070/aapl-20170930.xml'
# CIK code:320193
```

```python
response = requests.get(url)
```

```python
response.status_code
```

```python
print(response.text[:2000])
```

```python
import xml.etree.ElementTree as ET
```

```python
root = ET.fromstring(response.text)
```

```python
for child in root[:10]:
    print(child.tag)
```

```python
for child in root[:10]:
    print(child.attrib)
```

```python
import re
```

```python
[re.match('{(.*?)}', child.tag).group() for child in root]
```

```python
gaapSet = {}
for child in root:
    searchResult = re.match('{http://www.xbrl.org/2003/linkbase}(.*)', child.tag)
    if searchResult:
        gaapSet.add(searchResult.group(1))
```
