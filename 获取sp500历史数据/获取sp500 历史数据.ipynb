```python
%matplotlib qt
import datetime as dt
import matplotlib.pyplot as plt
from matplotlib import style
from mpl_finance import candlestick_ohlc
import matplotlib.dates as mdates
import pandas as pd
import pandas_datareader as web
import os
```

```python
start = dt.datetime(1990,1,1)
end = dt.datetime(2019,1,1)
```

```python
def download_and_save(stock):
    df = web.DataReader(stock, 'yahoo', start, end)
    df.to_csv('/Users/flu16/Desktop/sp500/%s.csv'%stock)
```

    获取sp500股票的symbol
   

```python
list_of_DataFrames = pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies', header=0)
```

```python
list_of_DataFrames[0].head()
```

```python
symbol_list = list(list_of_DataFrames[0].Symbol)
```

开始下载数据到桌面文件夹

```python
# 在桌面创建image文件夹
os.chdir('/Users/flu16/Desktop')
os.makedirs('sp500',exist_ok=True)
```

```python
for symbol in symbol_list:
    download_and_save(symbol)
    print('downloading stock %s'%(symbol))
```
